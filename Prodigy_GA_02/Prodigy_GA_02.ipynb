```python
import os
import torch
from diffusers import StableDiffusionPipeline
from IPython.display import display
```

```python
# Step 2: Load the Stable Diffusion model
model = "CompVis/stable-diffusion-v1-4"  # Use the official model
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
pipeline = StableDiffusionPipeline.from_pretrained(model, torch_dtype=torch.float16)
pipeline = pipeline.to(device)
```

```python
# Step 3: Generate an image
prompt = "Realistic Image of 2 armies of warriors using flowers fighting on a WarZone"
image = pipeline(prompt).images[0]
```

```python
output_dir = "generated_images"
os.makedirs(output_dir, exist_ok=True)
def get_next_image_number(directory):
    existing_files = os.listdir(directory)
    # Filter out files that match the pattern 'image<number>.png'
    image_files = [f for f in existing_files if f.startswith("image") and f.endswith(".png")]
    if not image_files:
        return 1  # If no images are present, start from 1
    # Extract numbers from filenames and find the maximum
    image_numbers = [int(f[5:-4]) for f in image_files]
    return max(image_numbers) + 1
```

```python
def generate_and_save_image(prompt):
    image = pipeline(prompt).images[0]
    # Get the next image number dynamically
    image_number = get_next_image_number("generated_images")
    # Dynamic filename
    filename = f"generated_images/image{image_number}.png"
    # Save the image
    image.save(filename)
    # Display the image inline
    display(image)
    print(f"Image saved as: {filename}")
```

```python
generate_and_save_image(prompt)
```
